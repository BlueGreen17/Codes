```python
#  All types of loops

# while loops
i = 0
while i < 10:
  print(i)
  i = i + 1

# for loops
for i in range(10):
  print(i)

# for loops with else
for i in range(10):
  print(i)
else:
  print("The loop is finished.")

# for loops with break
for i in range(10):
  if i == 5:
    break
  print(i)

# for loops with continue
for i in range(10):
  if i == 5:
    continue
  print(i)
```

```python
# Perform operations on array without using numpy

arr = [1, 2, 3, 4, 5]
print(arr[2])  # Access the element at index 2
print(arr[1:4])  # Access a range of elements
print(arr[-1])  # Access the last element
print(arr.count(3))  # Count the number of occurrences of an element
print(arr.index(3))  # Get the index of an element
arr.append(6)  # Append an element to the end of the array
arr.insert(2, 7)  # Insert an element at a specific index
arr.remove(3)  # Remove an element from the array
arr.pop()  # Remove the last element from the array
arr.sort()  # Sort the array in ascending order
arr.reverse()  # Reverse the order of the array
print(arr)
```

```python
# dictionary in python

dict1 = {
  "name": "Ananta",
  "age": 20,
  "city": "Gurgaon"
}
print(dict1)


print(dict1["name"])
print(dict1.get("age"))
print(dict1.get("country", "India"))


dict1["phone"] = "9811457704"
print(dict1)


dict1["age"] = 21
print(dict1)


del dict1["city"]
print(dict1)

# Iterating over a dictionary
for key, value in dict1.items():
  print(f"{key}: {value}")
```
